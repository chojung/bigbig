# Beautiful Soup

* pip install beautifulsoup4 -> 설치
    * 파서    |    설정
    * Python html.parser   |    BeautifulSoup(markup, "html.parser")
    * lxml parser	|    BeautifulSoup(markup, "lxml")

```python
from bs4 import BeautifulSoup
import requests

page = requests.get("https://www.python.org/")
soup = BeautifulSoup(page.content)

print soup.title
print soup.find_all('div',class_="do-not-print", id="top")
print "******************************************"
print soup.find_all('div',{'class': 'do-not-print','id':'top'})
```

```python
from bs4 import BeautifulSoup
import urllib

uResponse = urllib.urlopen("https://www.python.org/")
html=uResponse.read()

tree = BeautifulSoup(html,"lxml")

print tree.find_all('ul', class_="subnav menu", role="menu")
print "----------------------------"
strongtags = tree('ul', class_="subnav menu", role="menu")
for tag in strongtags:
    print tag
```

```python
from urllib import urlopen
from bs4 import BeautifulSoup

html=urlopen("http://en.wikipedia.org/wiki/HTML").read()
tree = BeautifulSoup(html,"lxml")

count=0
for link in tree.findAll('a'):
    if 'href' in link.attrs:
        if count<20:
            print count,link.attrs['href']
    count+=1
print "Total:",count
print"**************************************************"

for count,link in enumerate(tree.findAll('a')):
    if count<20:
        print count,link.get('href')
```

# Parsing
* 단계 1: 사용할 라이브러리를 준비한다.
    * import lxml.etree
* 단계 2: html 페이지를 가져온다 (메모리 또는 파일로 저장)
* 단계 3: 가져온 페이지를 파싱하고, 트리를 생성한다 (태그로 변환).
    * tree=lxml.etree.parse('my.html')
    * root=tree.getroot()
* 단계 4: 트리에서 필요한 요소 selector를 정한다.
* 단계 5: 필요한 요소를 가져온다.
    * root.getiterator():
        * 또는
    * tree.xpath() 또는 tree.css()

# Regex 정규식

```python
import re

sentence="Here goes my phone number 2287-1111. Nice to meet you! Merry Christmas <a>foo</a>"

regex1='\w+'
print "\w+ 숫자와 문자: ",re.findall(regex1, sentence),"\n"

regex2='\d+'
print "\d+ 숫자: ",re.findall(regex2, sentence),"\n"

regex3 = '[A-Z]\w+'
print "[A-Z]\w+ 대문자를 가진 단어: ",re.findall(regex3, sentence),"\n"

regex = "<a>(.*)</a>"
print "a태그안에 있는거 다 : ",re.findall(regex,sentence)
```

```python
import requests
import re

r=requests.get('http://python.org/')
html = r.text

p = re.compile('<h1>(.*?)</h1>')
tags = p.findall(html)

for tag in tags:
    print tag
```

* BeautifulSoup 에서 regex

```python
from bs4 import BeautifulSoup 
import re

htmlstr = """
<p>this is number 12345</p>
<p>this is text</p2>
"""

soup = BeautifulSoup(htmlstr)

for e in soup(text= re.compile('\d+')):
    print "tag:",e.parent
    print "text:",e
```

# CSS Select

```python
htmlstr="""<!DOCTYPE html>
<html>
<head>
<meta charset="UTF-8">
<title>My Home Page</title>
</head>
<body>
<h1>안녕하십니까</h1>
<p>오늘은 프로그래밍 하는 날...</p>
<p>Today we do programming...</p>
</body>
</html>"""

import lxml.html
from lxml.cssselect import CSSSelector

html = lxml.html.fromstring(htmlstr)
sel = CSSSelector('body p')
nodes = sel(html)
#print nodes
for node in nodes:
    print node.text
```

# 웹데이터-1: python.org 페이지를 크롤링해서 http url 출력하기

```python
import urllib2

url = 'http://python.org/'
request = urllib2.Request(url)
response = urllib2.urlopen(request)
html = response.read()

import re

p=re.compile('href="(http://.*?)"')
nodes = p.findall(html)

print "몇개 : ",len(nodes)

for i, node in enumerate(nodes):
    print i, node
```
